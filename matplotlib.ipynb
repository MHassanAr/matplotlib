```python
import pandas as pd  # Importing the pandas library for data manipulation
from collections import Counter  # Importing Counter from collections to count the languages

# Reading the CSV file containing the data and storing it in the 'data' DataFrame
data = pd.read_csv("data.csv")

# Extracting the 'Responder_id' column (not used further in this code)
ids = data["Responder_id"]

# Extracting the 'LanguagesWorkedWith' column, which contains the languages respondents worked with
lang = data["LanguagesWorkedWith"]

# Initializing a Counter to keep track of the number of people using each language
languageCounter = Counter()

# Iterating through each response (which contains a list of languages separated by semicolons)
for response in lang:
    # Splitting the languages by semicolon and updating the counter
    languageCounter.update(response.split(";"))

# Initializing empty lists to store languages and their popularity
language = []
popularity = []

# Getting the 15 most common languages and their corresponding counts
for item in languageCounter.most_common(15):
    # Appending the language to the 'language' list
    language.append(item[0])
    # Appending the number of users (popularity) to the 'popularity' list
    popularity.append(item[1])

# Reversing the order of the lists so the most popular languages appear at the top in the chart
language.reverse()
popularity.reverse()

# Plotting a horizontal bar graph with the languages on the y-axis and their popularity on the x-axis
plt.barh(language, popularity)

# Adding a label to the x-axis to indicate what the numbers represent
plt.xlabel("Number of people who use")

# Displaying the bar chart
plt.show()
```

<h3>Pie Chart</h3>

```python
# Data for the pie chart representing different values
slices = [60, 40, 30, 20]  

# Labels for each slice in the pie chart
labe = ["Sixty", "Forty", "Thirty", "Twenty"]  

# Custom colors for each slice in the pie chart
color = ["#CB6040", "#FD8B51", "#F2E5BF", "#257180"]  

# Creating the pie chart with data, labels, and colors
# 'wedgeprops' is used to customize the appearance of the pie chart's edges
plt.pie(slices, labels=labe, colors=color, wedgeprops={"edgecolor":"black"})

# Adding a title to the pie chart
plt.title("Pie Chart")

# Displaying the pie chart
plt.show()  # This displays the chart on the screen
```

```python
# Data for the popularity of programming languages
popularity = [59219, 55466, 47544, 36443, 35917]

# Labels for each programming language
languages = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# Custom colors for each slice of the pie chart
color = ["#86A7FC", "#B8001F", "#508C9B", "#DAD3BE", "#747264"]

# Exploding the 'Python' slice slightly to emphasize it
explode = [0, 0, 0, 0.1, 0]

# Creating the pie chart with data, labels, colors, explode effect, shadow, and percentage display
plt.pie(popularity, labels=languages, colors=color, explode=explode, 
        shadow=True, autopct="%1.1f%%", wedgeprops={"edgecolor":"black"})

# Adding a title to the pie chart
plt.title("Most popular programming languages")

# Displaying the pie chart
plt.show()  # This function call displays the chart on the screen
```
